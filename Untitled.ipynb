```python
correct = 0
wrong = 0
for idx, x in enumerate(latency_df['LogReturns'][:1825]):
    grad = 0
    if idx != 0:
        grad = latency_df['LogReturns'][idx] - latency_df['LogReturns'][idx-1]
        is_next_up = (latency_df['LogReturns'][idx+1] - latency_df['LogReturns'][idx]) > 0
        next_up_pred = grad > 0
        if is_next_up == next_up_pred:
            correct+=1
        else:
            wrong+=1
#     print(latency_df['LogReturns'][idx],latency_df['LogReturns'][idx+1])
```

```python
print(correct,wrong)
```

```python
fft_df = np.fft.fft(latency_df['LogReturns'][:256])
plt.plot(np.fft.fft(latency_df['LogReturns']))
```

```python
len(fft_df)
```

```python
res = np.zeros(len(latency_df['LogReturns']))
for k in range(0, len(latency_df['LogReturns'])):
    val = 0.0
    for n in range(0, len(fft_df)):
        # https://dsp.stackexchange.com/a/510/25943
        val += fft_df[n]*np.exp(1.j * 2*np.pi * n * k / len(fft_df)) / len(fft_df)
    res[k] = val.real

res = res[:len(latency_df['LogReturns'])]
#plot the result
plt.plot(res)
plt.show()
```

```python
len(latency_df['LogReturns'])
```

```python
pred_latency = [x>0 for x in res]
true_sign = [x>0 for x in latency_df['LogReturns']]
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(true_sign[256:],pred_latency[256:])
```
